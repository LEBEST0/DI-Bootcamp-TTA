# Exercice 1 : Chargement et exploration des données
Chargement du dataset Titanic depuis l'URL publique (chemin relatif évité pour la portabilité).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.impute import KNNImputer, SimpleImputer

# Chargement depuis l'URL publique pour assurer la portabilité du code
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_data = pd.read_csv(url)

print("Shape:", titanic_data.shape)
titanic_data.head()
```

```python
# Vérification des doublons
# On utilise titanic_data (et non data_ex1, variable non définie)
print('Nombre de lignes dupliquées :', titanic_data.duplicated().sum())
# Résultat : 0 → aucun doublon dans le dataset
```

# Exercice 2 : Gestion des valeurs manquantes
On traite les valeurs manquantes **sur le même dataframe** `titanic_data` afin de préserver le flux de travail cohérent.

```python
# Aperçu des valeurs manquantes
print(titanic_data.isnull().sum())
# → 177 valeurs manquantes dans Age, 687 dans Cabin, 2 dans Embarked
```

```python
# Suppression de la colonne Cabin : trop de valeurs manquantes (~77%)
# Conserver cette colonne introduirait plus de bruit que d'information utile
titanic_data = titanic_data.drop(columns=['Cabin'])
print('Cabin supprimée. Colonnes restantes :', list(titanic_data.columns))
```

```python
# Imputation de l'âge avec KNNImputer (k=2)
# KNN est préféré à la médiane car il prend en compte les corrélations entre Pclass, Age et Fare
imputer_knn = KNNImputer(n_neighbors=2)
imputed = imputer_knn.fit_transform(titanic_data[['Pclass', 'Age', 'Fare']])
titanic_data['Age'] = imputed[:, 1]

# Imputation de Embarked avec la valeur la plus fréquente (mode)
# Embarked est une variable catégorielle → stratégie 'most_frequent' appropriée
imputer_freq = SimpleImputer(strategy='most_frequent')
titanic_data['Embarked'] = imputer_freq.fit_transform(titanic_data[['Embarked']]).ravel()

print('Valeurs manquantes après imputation :')
print(titanic_data.isnull().sum())
```

# Exercice 3 : Ingénierie des fonctionnalités (Feature Engineering)
Création de nouvelles variables et encodage des variables catégorielles.
**Choix d'encodage :** On utilise `pd.get_dummies` (One-Hot Encoding) pour `Sex`, `Embarked` et `Title`.
Ce choix est cohérent avec l'exercice 6 et évite d'introduire un ordre arbitraire entre les catégories (ce que LabelEncoder ferait).

```python
# Création de la variable FamilySize
# SibSp = frères/soeurs/conjoints, Parch = parents/enfants, +1 pour le passager lui-même
titanic_data['FamilySize'] = titanic_data['SibSp'] + titanic_data['Parch'] + 1
print('FamilySize créée :', titanic_data['FamilySize'].value_counts().head())
```

```python
# Extraction du titre depuis le nom
# Ex : 'Braund, Mr. Owen Harris' → 'Mr'
titanic_data['Title'] = (
    titanic_data['Name']
    .str.split(',')
    .str[1]
    .str.split('.')
    .str[0]
    .str.strip()
)
print('Titres extraits :', titanic_data['Title'].value_counts())
```

```python
# One-Hot Encoding pour Sex, Embarked et Title
# Choix : get_dummies est préféré ici car ces variables n'ont pas d'ordre naturel
# drop_first=True évite la multi-colinéarité (dummy variable trap)
titanic_data = pd.get_dummies(titanic_data, columns=['Sex', 'Embarked', 'Title'], drop_first=True)

print('Colonnes après encodage :', list(titanic_data.columns))
titanic_data.head()
```

# Exercice 4 : Détection et traitement des valeurs aberrantes (Outliers)
On travaille sur le **même dataframe** `titanic_data` déjà préprocessé — **pas de rechargement du CSV**.
Méthode retenue pour traiter les outliers de `Fare` : **Quantile Capping (écrêtage)**,
car elle préserve toutes les lignes tout en limitant l'influence des valeurs extrêmes.

```python
# Visualisation avant traitement
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].boxplot(titanic_data['Fare'].dropna())
axes[0].set_title('Boxplot Fare (avant traitement)')
axes[1].hist(titanic_data['Fare'].dropna(), bins=30)
axes[1].set_title('Distribution Fare (avant traitement)')
plt.tight_layout()
plt.show()
```

```python
# Détection des outliers avec la méthode IQR pour Fare
Q1_fare = titanic_data['Fare'].quantile(0.25)
Q3_fare = titanic_data['Fare'].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
lower_bound = Q1_fare - 1.5 * IQR_fare
upper_bound = Q3_fare + 1.5 * IQR_fare

fare_outliers = titanic_data[(titanic_data['Fare'] < lower_bound) | (titanic_data['Fare'] > upper_bound)]
print(f'Borne basse : {lower_bound:.2f}, Borne haute : {upper_bound:.2f}')
print(f'Nombre d\'outliers détectés (Fare) : {fare_outliers.shape[0]}')
```

```python
# Traitement par Quantile Capping (écrêtage au 98e percentile)
# Cette méthode est choisie car elle conserve toutes les lignes
# et réduit l'influence des valeurs extrêmes sans les supprimer
upper_cap = titanic_data['Fare'].quantile(0.98)
titanic_data['Fare'] = titanic_data['Fare'].clip(upper=upper_cap)

# Vérification après traitement
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].boxplot(titanic_data['Fare'].dropna())
axes[0].set_title('Boxplot Fare (après écrêtage)')
axes[1].hist(titanic_data['Fare'].dropna(), bins=30)
axes[1].set_title('Distribution Fare (après écrêtage)')
plt.tight_layout()
plt.show()
print('Max Fare après traitement :', titanic_data['Fare'].max())
```

# Exercice 5 : Normalisation et standardisation
Le scaling est appliqué **après** le traitement des outliers (exercice 4), sur le même dataframe.
- `StandardScaler` pour `Age` et `FamilySize` : ces variables suivent approximativement une distribution normale.
- `MinMaxScaler` pour `Fare` : réduit la variable dans [0, 1] après l'écrêtage.

```python
standard_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()

# Avant scaling
print('Avant scaling :')
print(titanic_data[['Age', 'FamilySize', 'Fare']].describe())
```

```python
# StandardScaler sur Age et FamilySize
# Centrage-réduction : moyenne=0, écart-type=1
titanic_data[['Age', 'FamilySize']] = standard_scaler.fit_transform(titanic_data[['Age', 'FamilySize']])

# MinMaxScaler sur Fare
# Normalisation dans [0, 1] pour les algorithmes sensibles à l'échelle
titanic_data[['Fare']] = minmax_scaler.fit_transform(titanic_data[['Fare']])

print('Après scaling :')
print(titanic_data[['Age', 'FamilySize', 'Fare']].describe())
```

```python
# Visualisation après scaling
fig, axes = plt.subplots(1, 3, figsize=(15, 4))
axes[0].hist(titanic_data['Age'], bins=30)
axes[0].set_title('Age (standardisé)')
axes[1].hist(titanic_data['Fare'], bins=30)
axes[1].set_title('Fare (normalisé [0,1])')
axes[2].hist(titanic_data['FamilySize'], bins=30)
axes[2].set_title('FamilySize (standardisé)')
plt.tight_layout()
plt.show()
```

# Exercice 6 : Encodage des variables catégorielles
L'encodage One-Hot a déjà été appliqué sur `Sex`, `Embarked` et `Title` à l'**Exercice 3**,
dans le cadre d'un workflow cohérent et linéaire. Cette section affiche les résultats et les types de données actuels.

```python
# Vérification des types de données après tout le préprocessing
print('Types de données :')
print(titanic_data.dtypes)

# Vérification : plus aucune colonne object non encodée ne doit subsister
# (sauf Name et Ticket qui sont des identifiants)
remaining_obj = titanic_data.select_dtypes(include=['object']).columns.tolist()
print('\nColonnes de type object encore présentes :', remaining_obj)
```

```python
# Optionnel : suppression des colonnes identifiantes inutiles pour la modélisation
cols_to_drop = [col for col in ['Name', 'Ticket', 'PassengerId'] if col in titanic_data.columns]
titanic_data = titanic_data.drop(columns=cols_to_drop)
print('Colonnes supprimées :', cols_to_drop)
print('Shape finale :', titanic_data.shape)
titanic_data.head()
```

# Exercice 7 : Transformation de la variable Age (AgeGroup)
Création de groupes d'âge à partir de la variable `Age` **après scaling**.
Note : comme `Age` est maintenant standardisé, on utilise les quantiles pour définir les bins dynamiquement.

```python
# Puisque Age est standardisé (centrée-réduite), on utilise pd.qcut pour créer des groupes équilibrés
# Alternative : recréer AgeGroup avant le scaling (approche plus lisible pédagogiquement)
# Ici on utilise qcut pour s'adapter à la variable scaled
titanic_data['AgeGroup'] = pd.qcut(
    titanic_data['Age'],
    q=4,
    labels=['Q1_Young', 'Q2_YoungAdult', 'Q3_Adult', 'Q4_Senior']
)

print('Distribution des groupes d\'âge :')
print(titanic_data['AgeGroup'].value_counts())
```

```python
# One-Hot Encoding des groupes d'âge
# drop_first=True pour éviter la multicolinéarité
titanic_data = pd.get_dummies(titanic_data, columns=['AgeGroup'], drop_first=True)

# Vérification finale
print('Shape finale après toutes les transformations :', titanic_data.shape)
print('Colonnes finales :', list(titanic_data.columns))
titanic_data.head()
```
