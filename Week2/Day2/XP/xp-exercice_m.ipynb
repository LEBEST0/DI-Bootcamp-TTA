# Exercice 3 : Ingénierie des fonctionnalités

```python
import pandas as pd

titanic_data = pd.read_csv(r"C:\Users\AKRE\Documents\FORMATION\FORMATION TTA\DI-Bootcamp-TTA\Week2\Day2\XP\titanic dataset\train.csv")


# -----------------------------
# FamilySize feature
# -----------------------------
# SibSp = nombre de frères/soeurs/conjoints
# Parch = nombre de parents/enfants
# +1 pour compter le passager lui-même

titanic_data["FamilySize"] = (
    titanic_data["SibSp"]
    + titanic_data["Parch"]
    + 1
)

# =========================================
# 2. EXTRACT TITLE FROM NAME
# =========================================

# Exemple :
# "Braund, Mr. Owen Harris"
# -> "Mr"

titanic_data["Title"] = (
    titanic_data["Name"]
    .str.split(",")
    .str[1]
    .str.split(".")
    .str[0]
    .str.strip()
)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

titanic_data["Sex"] = encoder.fit_transform(
    titanic_data["Sex"]
)

titanic_data["Embarked"] = encoder.fit_transform(
    titanic_data["Embarked"]
)

titanic_data["Title"] = encoder.fit_transform(
    titanic_data["Title"]
)

print(
    titanic_data[
        ["Sex", "Embarked", "Title"]
    ].head()
)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic_data = pd.read_csv(
    r"C:\Users\AKRE\Documents\FORMATION\FORMATION TTA\DI-Bootcamp-TTA\Week2\Day2\XP\titanic dataset\train.csv"
)
```

```python
plt.boxplot(titanic_data["Fare"].dropna())
plt.title("Boxplot of Fare")
plt.show()
```

```python
plt.boxplot(titanic_data["Age"].dropna())
plt.title("Boxplot of Age")
plt.show()
```

```python
titanic_data["Fare"].hist(bins=30)
plt.title("Distribution of Fare")
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
titanic_data["Age"].hist(bins=30)
plt.title("Distribution of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
# Détection avec IQR

Q1 = titanic_data["Fare"].quantile(0.25)
Q3 = titanic_data["Fare"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print("Lower bound:", lower_bound)
print("Upper bound:", upper_bound)

fare_outliers = titanic_data[
    (titanic_data["Fare"] < lower_bound)
    |
    (titanic_data["Fare"] > upper_bound)
]

print(fare_outliers.shape)
print(fare_outliers[["Fare"]].head())
```

```python
# Méthode 1 : Quantile Capping
print(
    titanic_data["Fare"].quantile(0.98)
)

print(
    titanic_data["Fare"].quantile(0.99)
)


upper_cap = titanic_data["Fare"].quantile(0.98)

titanic_capped = titanic_data.copy()

titanic_capped["Fare"] = titanic_capped["Fare"].clip(
    upper=upper_cap
)
```

```python
# Vérifier après traitement
plt.boxplot(
    titanic_capped["Fare"].dropna()
)

plt.title("Fare After Quantile Capping")
plt.show()
```

```python
#Méthode  : Log Transformation


import numpy as np

titanic_log = titanic_data.copy()

titanic_log["Fare"] = np.log1p(
    titanic_log["Fare"]
)
```

```python
titanic_log["Fare"].hist(bins=30)

plt.title("Fare After Log Transformation")
plt.show()
```

```python
# mETHODE  : Row Removal

titanic_removed = titanic_data[
    (titanic_data["Fare"] >= lower_bound)
    &
    (titanic_data["Fare"] <= upper_bound)
]
```

```python
#Comparer le nombre de lignes

print("Before:", titanic_data.shape)

print("After removal:", titanic_removed.shape)
```
