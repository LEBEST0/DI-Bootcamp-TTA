# Exercise 5: Comparing Anxiety Levels Across Different Genders

```python
sns.barplot(x=df["Choose your gender"],
            y=df["Do you have Anxiety?"],
            hue=df["Choose your gender"],
            palette="Blues_d",
            legend=True)
plt.title("Comparing Anxiety Levels Across Different Genders")
plt.xlabel("Gender")
plt.ylabel("level of anxiety")
plt.ylim(0, 1)
plt.show()
```

# Exercise 6: Exploring the Relationship Between Age and Panic Attacks

```python
df["Do you have Panic attack?"]=df["Do you have Panic attack?"].map({"Yes":1,"No":0})
```

```python
df
```

```python
# Créer une NOUVELLE colonne Panic_num
df['Panic_num'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

plt.figure(figsize=(9, 5))
sns.scatterplot(
    data=df,
    x='Age',
    y='Panic_num',
    hue='Panic_num',                          
    palette={0: 'steelblue', 1: 'tomato'},    
    s=100,
    marker='o'
)

plt.title("Relation entre l'âge et les crises de panique")
plt.xlabel("Âge")
plt.ylabel("Crise de panique (1=Oui, 0=Non)")
plt.yticks([0, 1], ['Non', 'Oui'])
plt.legend(title='Crise de panique', labels=['Non', 'Oui'])
plt.show()
```
