```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
distribution=norm(50,10)
x_distribution=np.linspace(45,55,1000)
y_distribution=distribution.pdf(x_distribution)
plt.figure(figsize=(10,15))
plt.plot(y_distribution)
plt.title('Loi de gauss')
plt.xlabel('x')
plt.ylabel('valeurs de la distribution')
plt.show()
```

# Exercise 4: T-Test Application

```python
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

print(np.mean(data1))
print(np.mean(data2))
```

```python
t_stat,p_value=stats.ttest_ind(data1,data2)
print(f"Valeur de t_stat {t_stat}")
print(f"Valeur de p_value {p_value}")
```

# Exercise 5: Linear Regression Analysis

```python
# Given a dataset of housing prices (house_prices) and their corresponding sizes (house_sizes), use linear regression to predict the price of a house given its size.
house_sizes=np.array([50, 70, 80, 100, 120])
house_prices=np.array([150000, 200000, 210000, 250000, 280000])
# Performing linear regression
slope, intercept, r_value, p_value, std_err = stats.linregress(house_sizes,house_prices)
# The slope and intercept of the regression line
print("Slope:", slope)
print("Intercept:", intercept)


# Predict the price of a house that is 90 square meters.
print(f"price of of a house that is 90 square meters is : {slope*90+intercept}")
```

# Exercise 6: Understanding ANOVA

```python
fertilizer_1= [5, 6, 7, 6, 5]
fertilizer_2= [7, 8, 7, 9, 8]
fertilizer_3= [4, 5, 4, 3, 4]

f_value, p_value = stats.f_oneway(fertilizer_1, fertilizer_2, fertilizer_3)
print("F-value:", f_value)
print("P-value:", p_value)
```

# Based on the P-value, do the fertilizers have significantly different effects on plant growth?
let's defined Ho: all fertilizers have same effet
H1: there is one or more fertiliser that have different effect.
The result of ANOVA is 28.52 for f_value and 2.7542672098048448e-05 for p_value.
p_value is less than 0.05 so Ho is reject. It means that fertilizers have differents significants effects on plant growth

If the P-value were greater than 0.05, we can not reject Ho because we don't have enough statistic to justify that
